```python
import pandas as pd
from tqdm.auto import tqdm
from sqlalchemy import create_engine
```

```python
green_trips = pd.read_parquet("./green_tripdata_2025-11.parquet")
zones = pd.read_csv("./taxi_zone_lookup.csv")
```

```python
pg_user = 'root'
pg_password = 'root'
pg_host = 'localhost'
pg_port = '5432'
pg_db = 'ny_taxi'
target_table = 'green_taxi_data'
```
